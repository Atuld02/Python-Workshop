## Data Visualization
- An intresting way to display dataframe

```python
import matplotlib.pyplot as plt
a1=[1,2,3,4]
a2=[5,6,7,8]
plt.figure(figsize=(4,4))
plt.plot(a1,a2,color='red')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
